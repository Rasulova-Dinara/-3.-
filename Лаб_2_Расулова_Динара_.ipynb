```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt  # Импортируем библиотеку для визуализации

# Считывание данных
def read_data(file_path):
    data = pd.read_csv(file_path)
    return data

# Оценка линейной модели
def fit_model(X, y):
    X = sm.add_constant(X)  # добавление константы
    model = sm.OLS(y, X).fit()
    return model

# Отбор значимых факторов
def select_significant_factors(model, alpha=0.05):
    p_values = model.pvalues
    significant_factors = p_values[p_values < alpha].index.tolist()
    return significant_factors

# Оценка адекватности модели
def assess_model(model):
    print(model.summary())
    r_squared = model.rsquared
    f_statistic = model.fvalue
    return r_squared, f_statistic

# Вычисление RMSE
def calculate_rmse(y_true, y_pred):
    rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))
    return rmse

# Визуализация результатов
def plot_results(y_true, y_pred):
    plt.figure(figsize=(10, 5))
    plt.plot(y_true.index, y_true, label='Истинные значения', color='blue')
    plt.plot(y_pred.index, y_pred, label='Предсказанные значения', linestyle='--', color='orange')
    plt.legend()
    plt.title('Сравнение истинных и предсказанных значений')
    plt.xlabel('Наблюдения')
    plt.ylabel('MEDV (k$)')
    plt.show()

# Предсказание новых значений
def predict_new_values(model, new_data):
    new_data = sm.add_constant(new_data)
    predictions = model.predict(new_data)
    return predictions

# Основная функция
def main():
    # Запрос пути к файлу с новыми данными для предсказания в начале
    new_data_path = input("Введите путь к файлу с новыми данными для предсказания: ")

    # Загрузка данных
    data = read_data('boston.csv')
    y = data['MEDV']
    X = data.drop(columns=['MEDV'])

    # Оценка модели
    model = fit_model(X, y)

    # Отбор значимых факторов
    alpha = float(input("Введите уровень значимости (например, 0.05): "))
    significant_factors = select_significant_factors(model, alpha)
    print("Значимые факторы:", significant_factors)

    # Оценка адекватности модели
    r_squared, f_statistic = assess_model(model)
    print(f"Коэффициент детерминации: {r_squared}, F-статистика: {f_statistic}")

    # Предсказание
    y_pred = model.predict(sm.add_constant(X))

    # Вычисление RMSE
    rmse = calculate_rmse(y, y_pred)
    print(f"RMSE: {rmse}")

    # Визуализация
    plot_results(y, y_pred)

    # Сохранение вывода в текстовый файл
    with open('output.txt', 'w') as f:
        f.write("Значимые факторы:\n")
        f.write(str(significant_factors) + "\n")
        f.write(f"Коэффициент детерминации: {r_squared}\n")
        f.write(f"F-статистика: {f_statistic}\n")
        f.write(f"RMSE: {rmse}\n")

    # Пробел перед запросом на ввод новых данных
    print("\n")  # Добавляем пробел перед следующим вводом

    # Предсказание новых значений
    try:
        new_data = pd.read_csv(new_data_path)
        predictions = predict_new_values(model, new_data)
        print("Предсказанные значения:", predictions)
    except Exception as e:
        print(f"Ошибка при загрузке новых данных: {e}")

if __name__ == "__main__":  # Исправлено условие запуска основной функции
    main()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Считывание данных
def read_data(file_path):
    data = pd.read_csv(file_path)
    return data

# Создание временных лагов
def create_lags(data, lag_periods):
    for lag in lag_periods:
        for column in data.columns:
            data[f'{column}_lag_{lag}'] = data[column].shift(lag)
    return data.dropna()  # Удаляем строки с NaN после создания лагов

# Выделение подотрядов
def select_subseries(y, X, start, end):
    y_subseries = y[start:end]
    X_subseries = X.loc[y_subseries.index]  # Используем индексы y для X
    return y_subseries, X_subseries

# Оценка линейной модели
def fit_model(X, y):
    X = sm.add_constant(X)  # добавление константы
    model = sm.OLS(y, X).fit()
    return model

# Отбор значимых факторов
def select_significant_factors(model, alpha=0.05):
    p_values = model.pvalues
    significant_factors = p_values[p_values < alpha].index.tolist()
    return significant_factors

# Оценка адекватности модели
def assess_model(model):
    print(model.summary())
    r_squared = model.rsquared
    f_statistic = model.fvalue
    return r_squared, f_statistic

# Вычисление RMSE
def calculate_rmse(y_true, y_pred):
    rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))
    return rmse

# Визуализация результатов
def plot_results(y_true, y_pred):
    plt.figure(figsize=(10, 5))
    plt.plot(y_true.index, y_true, label='Истинные значения', color='blue')
    plt.plot(y_pred.index, y_pred, label='Предсказанные значения', linestyle='--', color='orange')
    plt.legend()
    plt.title('Сравнение истинных и предсказанных значений')
    plt.xlabel('Наблюдения')
    plt.ylabel('MEDV (k$)')
    plt.show()

# Основная функция
def main():
    # Загрузка данных
    data = read_data('boston.csv')
    y = data['MEDV']
    X = data.drop(columns=['MEDV'])

    # Параметры тюнинга
    lag_periods = [2]  # Пример лагов (1 и 2 периода)

    # Создание лагов для факторов
    X_with_lags = create_lags(X.copy(), lag_periods)

    # Подбор подотряда (например, с 10 по 15 наблюдение)
    start_index = 5
    end_index = 15
    y_subseries, X_subseries = select_subseries(y, X_with_lags, start_index, end_index)

    # Оценка модели
    model = fit_model(X_subseries, y_subseries)

    # Отбор значимых факторов
    alpha = float(input("Введите уровень значимости (например, 0.05): "))
    significant_factors = select_significant_factors(model, alpha)
    print("Значимые факторы:", significant_factors)

    # Оценка адекватности модели
    r_squared, f_statistic = assess_model(model)
    print(f"Коэффициент детерминации: {r_squared}, F-статистика: {f_statistic}")

    # Предсказание
    y_pred = model.predict(sm.add_constant(X_subseries))

    # Вычисление RMSE
    rmse = calculate_rmse(y_subseries, y_pred)
    print(f"RMSE: {rmse}")

    # Визуализация
    plot_results(y_subseries, y_pred)

    # Сохранение вывода в текстовый файл
    with open('output.txt', 'w') as f:
        f.write("Значимые факторы:\n")
        f.write(str(significant_factors) + "\n")
        f.write(f"Коэффициент детерминации: {r_squared}\n")
        f.write(f"F-статистика: {f_statistic}\n")
        f.write(f"RMSE: {rmse}\n")

if __name__ == "__main__":
    main()
```
